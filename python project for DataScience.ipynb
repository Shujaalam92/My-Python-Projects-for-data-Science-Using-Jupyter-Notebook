## Stock & Revenue Analysis: Tesla (TSLA) and GameStop (GME)
This notebook extracts stock data using `yfinance`, loads quarterly revenue from CSV files (fallback for scraping), and builds dashboards.

**Tip:** Run cells from top to bottom. This version avoids scraping errors by using public CSV datasets.

```python
# ==============================
# Setup
# ==============================
!pip install -q yfinance

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

%matplotlib inline
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 120)
```

## Q1 — Tesla Stock Data via yfinance (2 pts)

```python
tesla = yf.Ticker('TSLA')
tesla_stock_data = tesla.history(period='5y')
print('✅ Tesla Stock Data (first 5 rows):')
display(tesla_stock_data.head())
```

## Q2 — Tesla Quarterly Revenue (CSV Fallback, 1 pt)

```python
tesla_revenue = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/tesla-revenue.csv")
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace('[\$,]', '', regex=True).astype(float)
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

print("✅ Tesla Quarterly Revenue (last 10 rows):")
display(tesla_revenue.tail(10))
```

## Q3 — GameStop Stock Data via yfinance (2 pts)

```python
gme = yf.Ticker('GME')
gme_stock_data = gme.history(period='5y')
print('✅ GameStop Stock Data (first 5 rows):')
display(gme_stock_data.head())
```

## Q4 — GameStop Quarterly Revenue (CSV Fallback, 1 pt)

```python
gme_revenue = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/gamestop-revenue.csv")
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace('[\$,]', '', regex=True).astype(float)
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])

print("✅ GameStop Quarterly Revenue (last 10 rows):")
display(gme_revenue.tail(10))
```

## Q5 — Tesla Dashboard (Stock vs. Revenue) (2 pts)

```python
# Tesla Stock Price
plt.figure(figsize=(12, 6))
plt.plot(tesla_stock_data.index, tesla_stock_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# Tesla Quarterly Revenue
plt.figure(figsize=(12, 6))
plt.plot(tesla_revenue['Date'], tesla_revenue['Revenue'], label='Tesla Quarterly Revenue')
plt.title('Tesla Quarterly Revenue Over Time (USD Millions)')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions USD)')
plt.legend()
plt.show()
```

## Q6 — GameStop Dashboard (Stock vs. Revenue) (2 pts)

```python
# GameStop Stock Price
plt.figure(figsize=(12, 6))
plt.plot(gme_stock_data.index, gme_stock_data['Close'], label='GameStop Stock Price')
plt.title('GameStop Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# GameStop Quarterly Revenue
plt.figure(figsize=(12, 6))
plt.plot(gme_revenue['Date'], gme_revenue['Revenue'], label='GameStop Quarterly Revenue')
plt.title('GameStop Quarterly Revenue Over Time (USD Millions)')
plt.xlabel('Date')
plt.ylabel('Revenue (Millions USD)')
plt.legend()
plt.show()
```
